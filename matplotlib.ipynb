```python
categories = ['Electronics', 'Clothing', 'Books', 'Toys']
sales = [2000, 1500, 1000, 800]

import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.bar(categories,sales, color="skyblue", edgecolor="black")
plt.title("Product Category Sales")
plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.grid(axis="y")
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2400, 2000, 1800, 1500, 1200]

import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(months,sales, linestyle="--", color="red",marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
ages = [20, 25, 30, 35, 40, 45, 50, 55, 60]
frequency = [10, 15, 20, 25, 20, 15, 10, 5, 2]

data = []
for age, freq in zip(ages, frequency):
    data.extend([age] * freq) 
    
import matplotlib.pyplot as plt
plt.hist(data,bins=5,color="skyblue",edgecolor="black")
plt.title("Age Distibution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
time_periods = ['Q1', 'Q2', 'Q3', 'Q4']
product_A_sales = [100, 150, 200, 250]
product_B_sales = [80, 120, 180, 220]
product_C_sales = [50, 75, 125, 175]


import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(time_periods,product_A_sales,color="red",marker="o")
plt.plot(time_periods,product_B_sales, color="blue",marker="o")
plt.plot(time_periods,product_C_sales, color="Green",marker="o")
plt.title("Product Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
time = ['6 AM', '9 AM', '12 PM', '3 PM', '6 PM', '9 PM']
temperature = [20, 25, 30, 32, 28, 22]
humidity = [80, 75, 60, 55, 65, 75]

import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
plt.plot(time,temperature,color="blue",marker="o",label="Temperature")
plt.bar(time,humidity,color="skyblue",edgecolor="black",label="Humidity")
plt.title("Temperature and Humidity Trends")
plt.xlabel("Time")
plt.ylabel("Temperature / Humidity")
plt.legend(loc="upper right")  
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sales data for regions
regions = ['East', 'West', 'North', 'South']
sales = [1000, 800, 1200, 900]  # Sales values for each region

# Determine the region with the highest sales to explode its slice
explode = [0.1 if s == max(sales) else 0 for s in sales]

# Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(sales, labels=regions, explode=explode, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors[:len(regions)])
plt.title("Regional Sales Distribution")
plt.show()
```
