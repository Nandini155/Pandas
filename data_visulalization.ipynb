```python
# Challenge 1: Simple Line Plot
# Create a line plot for the following data:

# Requirements:
# Use a dashed red line.
# Add a title: "Negative Squares".
# Label the axes: "X Values" and "Y Values".
# Show the plot.

x = [0, 1, 2, 3, 4, 5]
y = [0, -1, -4, -9, -16, -25]

import matplotlib.pyplot as plt
plt.plot(x,y, linestyle="--", color="red")
plt.title("Negative Squares")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

```

```python
# Challenge 2: Multiple Line Plot
# Create a plot with two lines:

# Requirements:
# Add a legend to distinguish the lines.
# Add labels and a title: "Exponential Growth".


x = [0, 1, 2, 3, 4]
y1 = [1, 2, 4, 8, 16] #(use a blue line with circular markers)
y2 = [1, 3, 9, 27, 81] #(use a green dashed line with triangle markers)

plt.plot(x,y1, color="blue",marker="o",label="y = 2^x")
plt.plot(x,y2,color="green",linestyle="--",marker="^",label="y = 3^x")
plt.title("Exponential Growth")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# Challenge 3: Bar Chart
# Create a bar chart for the following data:
# Requirements:
# Use purple bars with black edges.
# Add a title: "Exam Scores".
# Label the axes: "Subjects" and "Scores".

Categories=['Math', 'Science', 'History', 'Art']
Scores=[85, 90, 78, 88]

plt.bar(Categories,Scores,color="purple", edgecolor="black",width=0.4)
plt.title("Exam Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
# Challenge 4: Scatter Plot
# Create a scatter plot for this data:

# Requirements:
# Use blue circles as markers.
# Add grid lines to the plot.
# Add labels and a title: "Random Data Points".

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 4, 1, 8, 3, 7, 5]

plt.scatter(x,y,color="blue",marker="o")
plt.title("Random Data Points")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
plt.show()
```

```python
#Question 1: Create a line chart showing the sales trends over 12 months.
# Use months as the x-axis (January to December).
# Assume sales data for each month is 
month=["January","February","March", "April","May","June","July","August","September","October","November","December"]
sales= [500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]

import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.plot(month,sales,color="Green", marker="o")
plt.title("Sales Trend over 12 Month")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
#Question 2: Create a bar chart comparing the sales data across the same 12 months.

month=["January","February","March", "April","May","June","July","August","September","October","November","December"]
sales= [500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
plt.figure(figsize=(15,8))
plt.bar(month,sales,color="blue",width=0.4)
plt.title("Sales Trend Over 12 Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()

```

```python
#Question 3: Create a scatter plot to compare sales vs. month.
month=["Jan","Feb","Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales= [500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
plt.figure(figsize=(10,6))
plt.scatter(month,sales,marker="o",color="orange")
plt.title("Sales Trend Over 12 Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()

average_sales= sum(sales)/len(sales)
plt.axhline(y=average_sales,color="red",linestyle="--",label=f'Average sales:{average_sales:.2f}')
plt.legend()
plt.show()
```

```python
# Question 4: Create a histogram to analyze the distribution of sales values.
# Use the same sales data.
# Divide the sales into 5 bins.
# Set the color of the bars to purple.
import pandas as pd
import matplotlib.pyplot as plt
data={
"month":["Jan","Feb","Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
"sales":[500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
}
df=pd.DataFrame(data)
df
plt.figure(figsize=(10,6))
plt.hist(x="sales",data=df,bins=5,color="Purple",edgecolor="black")
plt.title("Distribution of Sales Trend")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 5: Create a box plot to analyze the spread of sales data.
# Use the same sales data.
# Set the box color to cyan for a fresh look.
# Highlight the median value clearly.

import pandas as pd
import matplotlib.pyplot as plt
# data={
# "month":["Jan","Feb","Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
# "sales":[500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
# }
# df=pd.DataFrame(data)
# df
import numpy as np
month=["Jan","Feb","Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales= [500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
plt.figure(figsize=(6,6))
box=plt.boxplot(sales,patch_artist=True)
for element in ["boxes","whiskers", "caps","medians"]:
    plt.setp(box[element],color="cyan")
    plt.setp(box['medians'], color='red', linewidth=2)
plt.title("Box Plot of Monthly Sales")
plt.ylabel("Sales")
plt.show()
```

```python
#Question 6: Create a pie chart to show the contribution of each month's sales to the total.

import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(4,4))
data={
"month":["Jan","Feb","Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
"sales":[500, 700, 800, 600, 750, 900, 850, 1000, 1100, 950, 1200, 1300]
}
df=pd.DataFrame(data)
df
plt.pie(x=df["sales"],labels=df["month"],colors=plt.cm.Paired.colors,autopct="%1.2f%%")
plt.title("Monthly Sales Distribution")
plt.show()
```
